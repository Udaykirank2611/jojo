# **BFS**

```python
def bfs(graph,start,goal):
    visited = set()
    queue = [start]
    while queue:
        node = queue.pop(0)
        if node == goal:
            print("goal found")
            return True
        if node not in visited:
            visited.add(node)
            print("visited node : ",node)
            queue.extend(graph[node])
    print("goal not found")
    return False
```

# **DFS**

```python
def dfs(graph,start,goal):
    stack = [start]
    visited = set()
    while stack:
        node = stack.pop()
        if node == goal:
            print("goal found")
            return True
        if node not in visited:
            visited.add(node)
            print("visited node : ",node)
            stack.extend(reversed(graph[node]))
    print("goal not found")
    return False
```

# **GREEDY BEST FIRST SEARCH**

```python
import heapq
def gbfs(graph,start,goal,hea):
  visited = set()
  l = [(hea[start],start)]
  heapq.heapify(l)
  while l:
    c,cur = heapq.heappop(l)
    if cur==goal:
      print("Goal Found")
      return True
    if cur not in visited:
      visited.add(cur)
      print("visited node : ",cur)
      for neigh in graph[cur]:
        if neigh not in visited:
          heapq.heappush(l,(hea[neigh],neigh))
  print("goal not found")
  return False
```

# **A STAR**

```python
def astar(graph,costs,start,goal,hea):
  visited = set()
  l = [(0+hea[start],0,start,[start])]
  heapq.heapify(l)
  dic = {start:0}
  while l:
    f,cur_cost,node,path = heapq.heappop(l)
    if node == goal:
      print("goal found")
      print(path,cur_cost)
      return True
    for neigh in graph[node]:
      temp = cur_cost + costs[(node,neigh)]
      if neigh not in dic or temp < dic[neigh]:
        dic[neigh] = temp
        heapq.heappush(l,(temp+hea[neigh],temp,neigh,path+[neigh]))
  print("goal not found")
  return False
```

# **Linear Regression**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
weeks = np.array([2,3,4,5,6,7,8,9,10])
sales = np.array([5,6,10,12,11,14,18,20,22])
weeks_m = np.mean(weeks)
sales_m = np.mean(sales)
num = np.sum((weeks-weeks_m)*(sales-sales_m))
den = np.sum((weeks-weeks_m)**2)
b1 = num/den
b0 = sales_m - b1*weeks_m
pred = b0 + b1*11
plt.scatter(weeks,sales,label="Data")
plt.plot(weeks,b0+b1*weeks,label = 'Bestfit line',color ='red')
plt.xlabel("weeks")
plt.ylabel("sales")
plt.title("sales vs weeks")
plt.legend()
plt.show()
mae = np.mean(np.abs(sales-pred))
mse = np.mean((sales-pred)**2)
fmae = np.sqrt(mse)
print(mae,mse,fmae)
```
